```python
#Import statements
from pandas import * 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
```

```python
#Importing csv files
players = pd.read_csv("players.csv")
events = pd.read_csv("events.csv")
countries = pd.read_csv("countries.csv")
```

```python
american_players = players[players.country_id == 'USA'].sort(columns='name')
matches = american_players["name"].str.match(r'[A]')
index = 0
for match in matches:
    if(not match):
        break
    index += 1
american_players.head(index)
```

```python
grouping = players.groupby('country_id').size()

country_ids = []
num_players = []
for country in grouping.keys():
    country_ids.append(country)
    num_players.append(grouping[country])
    
DataFrame.from_dict({"country_id":country_ids,"num_players":num_players})
```

```python
merged = merge(countries, players, how='left', on='country_id')
merged = merged[merged.birthdate.str.match(r'1975')].groupby('country_id').size()

c_ids = []
num_1975 = []

for ids in countries.country_id:
    c_ids.append(ids)
    if(ids in merged):
        num_1975.append(merged[ids])
    else:
        num_1975.append(0)
        
DataFrame.from_dict({"country_id":c_ids,"num_players_born_1975":num_1975})
```
